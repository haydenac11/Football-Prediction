# This is the momentum stat that we defined. Ranges from -1 to 1 with -1 being bad, 0 being neutral, 1 being good

```python
# Now we want to generate the momentum row
def momentum(df, team, date):
    # filter the df by team
    team_filtered = df[(df['HomeTeam'] == team) | (df['AwayTeam'] == team)]

    # filter out games that happened after the date of game
    date_filtered = team_filtered[team_filtered['Date'] < date]

    # get the most recent 6 games
    recent_6 = date_filtered.sort_values(by='Date', ascending=False).head(6)

    #guard against empty df ie team has never been in league before
    # as it currently works, if a team has played less than 6 games for example 4 games then the momentum is calculated using the 4 games
    if len(recent_6) < 1:
        return 0

    # if team was not in league last season, set momentum to 0
    # if the most recent game was more than 300 days ago set to 0
    recent_game = recent_6['Date'].iloc[0]
    # print(recent_game)
    if (date - recent_game).days > 300:
    # if not wasPresentLastYear(df, team, date):
        print(f'Momentum date if statement')
        print(f'Team: {team}')
        print(f'Current Game Date: {date}')
        print(f'Most Recent Game Date: {recent_game}')
        print(f'Differential = {(date - recent_game).days}')
        return 0

    homewins = recent_6[(recent_6['HomeTeam'] == team) & (recent_6['FTR'] == 'H')]
    # print(len(homewins))
    awaywins = recent_6[(recent_6['AwayTeam'] == team) & (recent_6['FTR'] == 'A')]
    # print(awaywins)
    homeloss = recent_6[(recent_6['HomeTeam'] == team) & (recent_6['FTR'] == 'A')]
    # print(homeloss)
    awayloss = recent_6[(recent_6['AwayTeam'] == team) & (recent_6['FTR'] == 'H')]
    # print(awayloss)

    wins = len(homewins) + len(awaywins)
    losses = len(homeloss) + len(awayloss)

    m = (wins - losses) / len(recent_6)


    return m
#testing
# m = momentum(dirty_train, "Nott'm Forest", pd.to_datetime('2025-08-16'))
# print(m)
```

```python
home_momentum = []
away_momentum = []
for i in range(len(fresh_train)):
    date = fresh_train['Date'].iloc[i]
    HomeTeam = fresh_train['HomeTeam'].iloc[i]
    home_momentum.append(momentum(fresh_train, HomeTeam, date))
    AwayTeam = fresh_train['AwayTeam'].iloc[i]
    away_momentum.append(momentum(fresh_train, AwayTeam, date))

fresh_train.loc[:, 'HomeMomentum'] = home_momentum
fresh_train.loc[:, 'AwayMomentum'] = away_momentum

fresh_train.head()
```

## Function for calculating the win, loss, and draw percentage for a team over the past 38 games

```python

def wldPercentage(df, team, date):

    # filter games by our team of interest being home or away
    team_filtered = df[(df['HomeTeam'] == team) | (df['AwayTeam'] == team)]

    # filter out all the games that happen after this one (including this one)
    date_filtered = team_filtered[team_filtered['Date'] < date]

    # get most recent 38 games
    recent_38 = date_filtered.sort_values(by='Date', ascending=False).head(38)
    # print(recent_38)
    # guard against a divide by 0 error
    if len(recent_38) < 1:
        return 0, 0, 0

    # home_wins = len(recent_38[recent_38['FTR'] == 'H'])
    # home_loss = len(recent_38[recent_38['FTR'] == 'A'])
    # home_draw = len(recent_38[recent_38['FTR'] == 'D'])

    recent_game = recent_38['Date'].iloc[0]
    # print(recent_game)
    # math is just to account for maybe not having game in the first few weeks
    if (date - recent_game).days > (365 * 3) - 70:
        print(f'Momentum date if statement')
        print(f'Team: {team}')
        print(f'Current Game Date: {date}')
        print(f'Most Recent Game Date: {recent_game}')
        print(f'Differential = {(date - recent_game).days}')
        return 0, 1, 0

    homewins = len(recent_38[(recent_38['HomeTeam'] == team) & (recent_38['FTR'] == 'H')])
    # print(len(homewins))
    awaywins = len(recent_38[(recent_38['AwayTeam'] == team) & (recent_38['FTR'] == 'A')])
    # print(awaywins)
    homeloss = len(recent_38[(recent_38['HomeTeam'] == team) & (recent_38['FTR'] == 'A')])
    # print(homeloss)
    awayloss = len(recent_38[(recent_38['AwayTeam'] == team) & (recent_38['FTR'] == 'H')])

    homedraw = len(recent_38[(recent_38['HomeTeam'] == team) & (recent_38['FTR'] == 'D')])
    awaydraw = len(recent_38[(recent_38['AwayTeam'] == team) & (recent_38['FTR'] == 'D')])

    win_percentage = (homewins + awaywins) / len(recent_38)
    loss_percentage = (homeloss + awayloss) / len(recent_38)
    draw_percentage = (homedraw + awaydraw) / len(recent_38)


    return win_percentage, loss_percentage, draw_percentage

# testing
# dirty_train.head()
tw,td,tl = wldPercentage(dirty_train, "Arsenal", pd.to_datetime('2024-12-16'))
print(tw, td, tl)
```

### This function find the last time the 2 teams played eachother and finds who won. The returns are defined in the top of the function

```python
def HVA(df, TeamA, TeamB, date):
    # represents a draw as 0, win as 1, loss as -1
    # returns 2 numbers: Astat, Bstat
    # Get all the games the two teams have played against eachother
    team_filtered = df[
        ((df['HomeTeam'] == TeamA) & (df['AwayTeam'] == TeamB)) |
        ((df['HomeTeam'] == TeamB) & (df['AwayTeam'] == TeamA))
    ]
    # Case 1: The teams have never played before:
    date_filtered = team_filtered[(team_filtered['Date'] < date)]

    if len(date_filtered) < 1:
        return 0, 0


    most_recent = date_filtered.sort_values(by='Date', ascending=False).iloc[0]

    recent_game = most_recent['Date']

    # Case 1.5: Last game was more than 3 years ago
    if (date - recent_game).days > (365 * 3) - 70:
        print(f'More than 3 years if statement')
        print(f'Team: {TeamA} {TeamB}')
        print(f'Current Game Date: {date}')
        print(f'Most Recent Game Date: {recent_game}')
        print(f'Differential = {(date - recent_game).days}')
        return 0, 0


    # Case 2: TeamA won last game
    if most_recent['FTR'] == 'H' and most_recent['HomeTeam'] == TeamA:
        return 1, -1
    elif most_recent['FTR'] == 'A' and most_recent['AwayTeam'] == TeamA:
        return 1, -1
    # Case 3: Draw Last game
    elif most_recent['FTR'] == 'D':
        return 0, 0
    #Case 4: Team A lost last game
    else:
        return -1, 1

    print('Something bad happened')
    return
```

## This function finds the median value of a stat for a team over the past 6 games.

```python
def previousStats(df, team, date):
    stats_to_check = ['HS', 'AS', 'HST', 'AST', 'HF', 'AF', 'HC', 'AC', 'HY', 'AY', 'HR', 'AR']

    team_filtered = df[(df['HomeTeam'] == team) | (df['AwayTeam'] == team)]

    date_filtered = team_filtered[(team_filtered['Date'] < date)]
    recent_6 = date_filtered.sort_values(by='Date', ascending=False).head(6)

    if len(recent_6) < 1:
        return np.zeros_like(np.array(stats_to_check), dtype='float64')

    recent_game = recent_6['Date'].iloc[0]
    # print(recent_game)
    if (date - recent_game).days > 300:
    # if not wasPresentLastYear(df, team, date):
        print(f'PreviousStats if statement')
        print(f'Team: {team}')
        print(f'Current Game Date: {date}')
        print(f'Most Recent Game Date: {recent_game}')
        print(f'Differential = {(date - recent_game).days}')
        return np.zeros_like(np.array(stats_to_check), dtype='float64')

    arranged_columns = []
    for i in range(len(recent_6)):
        if recent_6['HomeTeam'].iloc[i] == team:
            arranged_columns.append(list(recent_6[stats_to_check].iloc[i]))
        else: # The team was the away team
            disorg = list(recent_6[stats_to_check].iloc[i])
            # swap the H and A columns and append
            arranged_columns.append([x for pair in zip(disorg[1::2], disorg[::2]) for x in pair])

    medians = np.median(arranged_columns, axis=0)

    return medians

```

```python
home_team_hva = []
away_team_hva = []

for i in range(len(fresh_train)):
    date = fresh_train['Date'].iloc[i]
    home_team = fresh_train['HomeTeam'].iloc[i]
    away_team = fresh_train['AwayTeam'].iloc[i]
    h, a = HVA(fresh_train, home_team, away_team, date)
    home_team_hva.append(h)
    away_team_hva.append(a)


fresh_train.loc[:, 'HomeHVA'] = home_team_hva
fresh_train.loc[:, 'AwayHVA'] = away_team_hva

fresh_train.tail()
```

```python
home_prev = []
away_prev = []

for i in range(len(fresh_train)):
    date = fresh_train['Date'].iloc[i]
    home_team = fresh_train['HomeTeam'].iloc[i]
    away_team = fresh_train['AwayTeam'].iloc[i]

    home_medians = previousStats(dirty_train, home_team, date)
    home_prev.append(home_medians)
    away_medians = previousStats(dirty_train, home_team, date)
    away_prev.append(away_medians)

```
