```python
import pandas as pd
```

```python
season23_24 = pd.read_csv("season23-24.csv")
season23_24 = season23_24.iloc[:, :24]
season23_24.head()
```

```python
season23_24.info()
```

```python
percent_missing =round(100*(season23_24.isnull().sum())/len(season23_24),2)
print(percent_missing)
```

```python
import matplotlib.pyplot as plt
colors = ['blue', 'red', 'green']
# Example: Plot a bar chart for column 'column_name'
season23_24['FTR'].value_counts().plot(kind='bar', color=colors)

plt.xlabel('Home team win, away team win, draws')  # Label for x-axis
plt.ylabel('Count')       # Label for y-axis
plt.show()
```

```python
# Filter only home wins
home_wins = season23_24[season23_24['FTR'] == 'H']

# Count the number of home wins per HS value
wins_per_shots = home_wins['HS'].value_counts().sort_index()

# Plot
plt.bar(wins_per_shots.index, wins_per_shots.values, color='blue')

# Add labels
plt.xlabel('Home Team Shots (HS)')
plt.ylabel('Number of Wins')
plt.title('Home Wins vs. Home Team Shots')
plt.xticks(range(0, season23_24['HS'].max() + 1, 2))  # Adjust x-axis for readability

# Annotate bars with values
for i, value in enumerate(wins_per_shots.values):
    plt.text(wins_per_shots.index[i], value + 0.2, str(value), ha='center', fontsize=10, fontweight='bold')

plt.show()
```
